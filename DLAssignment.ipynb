```python

import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_derivative(x):
    return x * (1 - x)
def relu(x):
    return np.maximum(0, x)
def relu_derivative(x):
    return np.where(x > 0, 1, 0)

X = np.array([[0], [1], [2], [3], [4]], dtype=float)
y = np.array([[0], [2], [4], [6], [8]], dtype=float)

np.random.seed(42)

W1 = np.random.randn(1, 3)
b1 = np.zeros((1, 3))

W2 = np.random.randn(3, 1)
b2 = np.zeros((1, 1))

learning_rate = 0.01
epochs = 1000
activation = "relu"
for epoch in range(epochs):
    z1 = np.dot(X, W1) + b1

    if activation == "sigmoid":
        a1 = sigmoid(z1)
    else:
        a1 = relu(z1)
    z2 = np.dot(a1, W2) + b2
    y_pred = z2
    loss = np.mean((y - y_pred) ** 2)

    d_loss = 2 * (y_pred - y) / y.shape[0]

    dW2 = np.dot(a1.T, d_loss)
    db2 = np.sum(d_loss, axis=0, keepdims=True)

    if activation == "sigmoid":
        d_hidden = np.dot(d_loss, W2.T) * sigmoid_derivative(a1)
    else:
        d_hidden = np.dot(d_loss, W2.T) * relu_derivative(a1)

    dW1 = np.dot(X.T, d_hidden)
    db1 = np.sum(d_hidden, axis=0, keepdims=True)

    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

print("\nFinal Predictions:")
print(y_pred)
```

Gradients become very small → slow learning

**ReLU avoids vanishing gradient bold text**

Gradient = 1 for positive values → faster updates

# epoches and learning rates comparision

Understand how learning rate affects convergence

Understand how epochs affect learning quality

```python
import numpy as np
import matplotlib.pyplot as plt
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

X = np.array([[0], [1], [2], [3], [4]], dtype=float)
y = np.array([[0], [2], [4], [6], [8]], dtype=float)


learning_rates = [0.00001, 0.0001, 0.001, 0.01, 0.1, 1]

epochs = 2000

loss_history_all = {}

for lr in learning_rates:
    np.random.seed(42)
    W1 = np.random.randn(1, 3) * 0.01
    b1 = np.zeros((1, 3))
    W2 = np.random.randn(3, 1) * 0.01
    b2 = np.zeros((1, 1))

    loss_history = []

    for epoch in range(epochs):
        z1 = np.dot(X, W1) + b1
        a1 = relu(z1)

        z2 = np.dot(a1, W2) + b2
        y_pred = z2


        loss = np.mean((y - y_pred) ** 2)
        loss_history.append(loss)
        d_loss = 2 * (y_pred - y) / y.shape[0]

        dW2 = np.dot(a1.T, d_loss)
        db2 = np.sum(d_loss, axis=0, keepdims=True)

        d_hidden = np.dot(d_loss, W2.T) * relu_derivative(a1)

        dW1 = np.dot(X.T, d_hidden)
        db1 = np.sum(d_hidden, axis=0, keepdims=True)


        W1 -= lr * dW1
        b1 -= lr * db1
        W2 -= lr * dW2
        b2 -= lr * db2

    loss_history_all[lr] = loss_history

    print(f"Learning Rate {lr} → Final Loss: {loss:.4f}")

for lr in learning_rates:
    plt.plot(loss_history_all[lr], label=f"lr={lr}")

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss vs Epochs for Different Learning Rates")
plt.legend()
plt.show()
```

#Perceptron (Step vs Sigmoid)
Understand how a single neuron behaves

Comparring:
Step function (classical perceptron)

Sigmoid (modern neural networks)

```python
import numpy as np

def step_function(x):
    return np.where(x >= 0, 1, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([[0], [1], [1], [0]])


W = np.array([[0.5], [0.5]])
b = -0.7


z = np.dot(X, W) + b

output_step = step_function(z)
output_sigmoid = sigmoid(z)


print("Input:\n", X)
print("\nActual Output:\n", y)

print("\nStep Function Output:\n", output_step)
print("\nSigmoid Output:\n", output_sigmoid)
```

**Step Function vs Sigmoid**

Step Output: Only 0 or 1

Sigmoid Output: Values like 0.33, 0.45, 0.57

Step Function

Produces hard binary output (0 or 1)

Decision is abrupt (sudden jump)

Output jumps directly from 0 → 1

Sigmoid Function

Produces smooth continuous output (0 to 1)

Represents probability

Example:

0.45 → “somewhat 0”
0.57 → “somewhat 1”

Sigmoid gives confidence level, Step gives final decision

**Effect of Bias**

Bias = -0.7

Correct classification

Bias = +0.7

All outputs become 1

Observations:

Bias shifts decision boundary

Negative bias → stricter condition

Positive bias → easier activation


Decision rule:
z = w1*x1 + w2*x2 + b

If b is large → z always positive → output = 1

If b is small → harder to reach threshold

Bias controls how easily neuron activates

**Effect of Weights**

With [0.5, 0.5] → correct AND behavior

Weights define importance of inputs

Higher weight → more influence

Lower weight → less influence

Decision boundary depends on weights

**XOR Failure**

Input	Expected	Predicted

(0,1)	1	0

(1,0)	1	0

Perceptron fails to learn XOR

XOR is NOT linearly separable

Meaning:

You cannot draw a single straight line to separate classes


Single perceptron = linear model
Cannot solve non-linear problems like XOR

**XOR problem led to invention of Multi-layer networks (MLP)**
