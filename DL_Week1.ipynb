```python
import torch
import torch.nn as nn
import torch.optim as optim

x=torch.randn(100,4)
y=torch.randn(100,1)

class SimpleNetwork(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear=nn.Linear(4,1)
  def forward(self,x):
    return self.linear(x)

model=SimpleNetwork()
loss_fn=nn.MSELoss()
optimizer=optim.Adam(model.parameters(),lr=0.5)

for epoch in range(100):
  y_pred=model(x)
  loss=loss_fn(y_pred,y)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

print("final loss:",loss.item())


```

```python
import tensorflow as tf

x=tf.random.normal((100,3))
y=tf.random.normal((100,1))
model=tf.keras.Sequential([tf.keras.layers.Dense(1,input_shape=(3,))])
model.compile(optimizer='adam',loss='mse')
model.fit(x,y,epochs=100,verbose=0)

print("final loss:",model.evaluate(x,y))
```

```python
from tensorflow import keras
from tensorflow.keras import layers

x=tf.random.normal((100,3))
y=tf.random.normal((100,1))

model=keras.Sequential([layers.Dense(1,input_shape=(3,))])
model.compile(optimizer='adam',loss='mse')
model.fit(x,y,epochs=120,verbose=0)
print("final loss:",model.evaluate(x,y))
```

```python
import numpy as np
def step(x):
  if x>=0:
    return 1
  else:
    return 0
class Perceptron:
  def __init__(self,weights,bias):
    self.weights=weights
    self.bias=bias
  def predict(self,inputs):
    result=np.dot(self.weights,inputs)+self.bias
    return step(result)

weights=np.array([1,1])
bias=-1.5

and_gate=Perceptron(weights,bias)

print("and gate")
for input in [(0,0),(0,1),(1,0),(1,1)]:
  print(input,' -> ',and_gate.predict(np.array(input)))
```
